## 🏥 1. Business Problem
##### - Diabetes is a growing health concern worldwide.
##### - Early detection helps in timely medical intervention and lifestyle changes.

## 💡 Objective:
##### - Use patient health data to predict whether a person is likely to be dibetic.
##### - I'll use KNN as our model - a simple and and interpretable algorithm that classifies new data based on similar past patients.

## Dataset Overview
##### I am using Pima Indians diabetes dataset from the UCI repository (via sklearn.dataset or CSV).
##### This data set was collected by a national Institute of diabetes and digestive and kidney diseases.
##### - 768 samples.
##### - Binary target: diabetic (1) or Not (0)

## feature descriptions
##### Pregnancies - number of times, pregnant
##### Glucose - Plasma glucose concentration
##### BloodPressure - diastolic blood pressure (mm Hg)
##### SkinThickness - triceps skin fold thickness (mm)
##### Insulin - 2-Hour serum insulin (mu U/ml)
##### BMI - Body Mass Index
##### DiabetesPedigreeFunction - Diabetes Pedigree function
##### Age - Age(in years)
##### Outcome - 1 = diabetic, 0 = Non diabetic

## 6. Feature Analysis

```python
# # split into train and test
# X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
X_train, X_test, y_train, y_test = preprocess_data(df)
X_train.shape, y_train.shape
```

## 7. Model Training

```python
best_model, best_k = tune_knn(X_train, y_train)
print("Best k:", best_k)
```

```python
# Train and Evaluate

model = KNN(k=7)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

## 8. Model Evaluation

```python
accuracy, matrix, report = evaluate_model(best_model, X_test, y_test)

print(f"Accuracy:, {accuracy:.2f}\n")
print("Confusion Matrix:\n",matrix)
print("\nReport:\n",report)
```

## 9. Data Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(matrix, annot=True, fmt='d', cmap='coolwarm')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

## 10. Conclusion
#### The KNN model achieved an accuracy of 71% after tuning k to 11, improving from 68%. The confusion matrix shows that the model performs better in predicting non-diabetic cases than diabetic ones. This is supported by higher precision (0.76) and recall (0.79) for class 0 compared to lower values for class 1 (precision 0.60, recall 0.56). The model struggles to correctly identify diabetic patients, leading to more false negatives. Overall, the model performs moderately well but can be improved further.
